# e-Statからダウンロードしたデータから演習用csvを出力する
- ラベル名を日本語から英語にするついでに， yyyymm形式のカラムもつける

```python
import pandas as pd
```

```python
df = pd.read_csv("FEH_00200561_200619194455.csv", encoding='cp932', skiprows=11)
df
```

```python
labels = pd.read_excel("label_info.xlsx", index_col=0)
labels.head()
```

```python
df_conv = df.rename(columns=labels.to_dict()["English"]).rename(columns={"時間軸（月次）":"time"})
df_conv.head()
```

```python
df_conv = df_conv[["time"] +  list(labels["English"])]
```

```python
df_conv["time"] = pd.to_datetime(df_conv["time"], format="%Y年%m月")
```

```python
df_conv["year"] = df_conv.apply(lambda x: x["time"].year, axis=1)
df_conv["month"] = df_conv.apply(lambda x: x["time"].month, axis=1)
```

```python
df = df_conv[["time", "year", "month"] +  list(labels["English"])]
```

```python
df.to_csv("week08data.csv", index=False)
```
